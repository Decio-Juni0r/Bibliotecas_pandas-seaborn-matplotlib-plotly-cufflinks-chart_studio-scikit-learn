```
#O método corr() do DataFrame base_casa para calcular a matriz de correlação entre as colunas numéricas do DataFrame.
#Correlação é uma tabela que mostra os coeficientes de correlação entre todos os pares de colunas numéricas em um DataFrame.
#O coeficiente de correlação mede a força e a direção da relação linear entre duas variáveis.
#Ele varia de -1 a +1:
#+1 indica uma correlação positiva perfeita (as variáveis aumentam juntas).
#-1 indica uma correlação negativa perfeita (uma variável aumenta enquanto a outra diminui).
#0 indica que não há correlação linear entre as variáveis.

base_casa.corr()
```

```
#Preparando os dados para serem usados em um modelo de Machine Learning. O modelo será treinado para aprender a relação entre as variáveis independentes (X)
#e a variável dependente (Y), para que possa prever o preço do imóvel (Y) com base nas características do imóvel e da área (X).
x = base_casa[['Avg_Area_Income', 'Avg_Area_House_Age', 'Avg_Area_Number_Rooms', 'Avg_Area_Number_of_Bedrooms', 'Area_Population']]
y = base_casa[['Price']]
```

```
from sklearn.model_selection import train_test_split
#from sklearn.model_selection import train_test_split: Importa a função train_test_split do módulo model_selection da biblioteca
#scikit-learn. Esta função é usada para dividir os dados em conjuntos de treinamento e teste.

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.7, test_size = 0.3, random_state = 42)
#x_train, x_test, y_train, y_test =: Atribui os resultados da função train_test_split a quatro variáveis:
#x_train: Conjunto de treinamento para as variáveis independentes (features).
#x_test: Conjunto de teste para as variáveis independentes (features).
#y_train: Conjunto de treinamento para a variável dependente (target).
#y_test: Conjunto de teste para a variável dependente (target).
#train_test_split(x, y, ...): Chama a função train_test_split com os seguintes argumentos:
#x: DataFrame contendo as variáveis independentes (features).
#y: DataFrame contendo a variável dependente (target).
#train_size = 0.7: Define o tamanho do conjunto de treinamento como 70% dos dados.
#test_size = 0.3: Define o tamanho do conjunto de teste como 30% dos dados.
#random_state = 42: Define uma semente aleatória para garantir que a divisão dos dados seja reproduzível.
```

```
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
#Exibindo as dimensões dos conjuntos de dados de treinamento e teste, permitindo que verifique se a divisão dos dados
#foi feita corretamente e se os conjuntos têm o número esperado de linhas e colunas.
```

```
from sklearn.linear_model import LinearRegression
#importando a classe LinearRegression da biblioteca scikit-learn, para que use essa classe para criar e treinar modelos de regressão linear
```

```
lm = LinearRegression()
#Criando um objeto do modelo de regressão linear que pode ser usado para treinar o modelo e fazer previsões em novos dados.
#O objeto lm é uma instância da classe LinearRegression.
```

```
lm.fit(x_train, y_train)
#Treinando o modelo de regressão linear lm usando os dados de treinamento x_train e y_train.
#O modelo aprenderá a relação entre as variáveis independentes e a variável dependente durante o processo de treinamento.
```

```
y_pred = lm.predict(x_test)
#Usando o modelo de regressão linear treinado lm para fazer previsões sobre os dados de teste x_test e armazena as previsões na variável y_pred.
#As previsões são os valores que o modelo estima para a variável dependente (target) com base nos valores das variáveis independentes (features) nos dados de teste.
```

```
from sklearn.metrics import r2_score
r = r2_score(y_test, y_pred)
#Calcula o R² do modelo de regressão linear usando os valores reais (y_test) e os valores previstos (y_pred) nos dados de teste.
#O R² é uma medida estatística que indica a proporção da variância na variável dependente que é explicada pelas variáveis independentes no modelo.
```

```
print("r_quadrado: ", r)
#O valor do R² na saída, precedido pela string "r_quadrado: ".
#Isso permite que você visualize o valor do R² e avalie o desempenho do modelo de regressão linear.
```
